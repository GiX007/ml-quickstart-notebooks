All we need is a dictionary with bases as keys and the their frequency of appearance as their values. We also need to make this calculation for 2 strings, so a function will help us to avoid repetition. It takes as input a string and outputs the bases and errors in dictionaries.

````python
# count all bases and errors in a DNA sequence
def count_all_bases_errors(dna_string):
    # initialize a dictionary-counter for bases and errors
    base_error_counts = {"A": 0, "G": 0, "T": 0, "C": 0, "-": 0, "N": 0}
    # iterate over the dictionary
    for base in base_error_counts.keys():
        base_error_counts[base] += dna_string.count(base) # for each base count how many times it appears in string and update the dictionary
    # bases and errors in seperate dictionaries
    bases = {base: count for base, count in base_error_counts.items() if base not in ["-", "N"]}
    errors = {base: count for base, count in base_error_counts.items() if base in ["-", "N"]}
    return bases, errors
````

Now, we are going to call the above function for the two given strings and print the resuts in the appropriate form.

````python
# Count bases and errors in the first string and print the results
counts_string_1 = count_all_bases_errors(string_1)
print( f"TOTAL BASES FOR STRING 1: {sum(counts_string_1[0].values())}, ERRORS: {sum(counts_string_1[1].values())}, {counts_string_1[0]}")

# Count bases and errors in the second string and print the results
counts_string_2 = count_all_bases_errors(string_2)
print(f"TOTAL BASES FOR STRING 2: {sum(counts_string_2[0].values())}, ERRORS: {sum(counts_string_2[1].values())}, {counts_string_2[0]}")
````

### **Exercise 2**

Calculate the number of matches (per base) and mismatches, when you align the strings. E.g.,you can already see 3 matches at positions 5, 8, 9 (G<-->G, G<-->G, T<-->T), and 7 mismatches(-<-->-, C<-->-, A<-->C, G<-->A, etc) in the left area of the two strings:

-CAGGTGGT... (first string)

--CAGGTGT... (second string)


An example of the output should be like: A-A matches=10, G-G matches=20, T-T matches=15,C-C matches=5, mismatches=100.

We will use another function here to calculate all matches and mismatches. It takes the two given strings as input and outputs how many matches and mismatches exist

````python
def count_matches_and_mismatches(string_1, string_2):
    # initialize a dictionary-counter for the matches
    matches = {'A': 0, 'G': 0, 'T': 0, 'C': 0}
    mismatches = 0

    # iterate through the paired input strings
    for base1, base2 in zip(string_1, string_2):
        # find matches and mismatches
        if base1 == base2:
            if base1 in matches:
                matches[base1] += 1 # count matches for all bases
        else:
            mismatches += 1  # count mismatches

    return matches, mismatches
````

We call the function using the given strings as inouts and print the results.

````python
# find matches and mismatches
matches, mismatches = count_matches_and_mismatches(string_1, string_2)

# print the results in the appropriate form
for base, count in matches.items():
    print(f"{base}-{base} matches = {count}")
print(f"mismatches = {mismatches}")
````

The PFM is a matrix of the form:

The values in the cells are the frequencies of bases' occurence in strings' positions. E.g., 11 means that "A" appears 11 times in the 1st position of each string, 0 means that that "A" appears 0 times in the 2nd/4th position of each string, etc. An example of the output should be like:

First, we create a list with all given dna sentences.

````python
# dna strings in a list
dna_string_list = ["AGATAG", "AGATAG", "AGATAG", "AGATAG", "CGATAG", "AGATAC", "AGATAA", "GGATAC", "CGATTG", "CGATTG",
                   "AGATAG", "AGAGAG", "CGATAC", "AGATTG", "CGATAG", "AGATAC", "AGATAA", "GGATAC", "CGATTG", "CGATTG" ]
````

````python
# print the length of the list and the length of each element
print(f"The length of dna_sequences is: {len(dna_string_list)}")
print(f"The length of the first element is: {len(dna_string_list[0])}")
````

All the given dna sentences have the same length, which is 6. We create a dictionary with keys the bases and values lists for storing the frequency at each position. Then, a function with the given string list as input will loop over the elements of the list and over the length of each element in order to count the occurence of each base at each position and store it to the dictionary, which is its output. So, we calculate the PFM and print the results.

````python
# all elements of the list have the same length
string_length = len(dna_string_list[0])

# create a dictionary for storing the frequency counts for all bases
pfm = {
    'A': [0] * string_length,
    'T': [0] * string_length,
    'G': [0] * string_length,
    'C': [0] * string_length
}

# function that calculates the pfm
def count_pfm(strings_list):
  # loop over the list
  for seq in strings_list:
    # loop over the string
    for i, base in enumerate(seq):
      # count the occurrence of each base at each position
      pfm[base][i] += 1
  return pfm

# calculate pfm
pfm = count_pfm(dna_string_list)

# create a list of bases that helps us in the printing loop
bases = ['A', 'T', 'G', 'C']

# print the Position Frequency Matrix (PFM)
print("Position Frequency Matrix:")
for base in bases:
  # modify the print to be without brackets
  print(f"{base}: {', '.join(map(str, pfm[base]))}")  # for array of integer type, we need to change it to string type first and than use join function to get clean output without brackets
````

To begin with, we need to create 2 files that contain a list(lines) of lists(each line is a list of 5 random dna strings of a X length). This will happen by using some simple functions, based on the `00_python_basics.ipynb`.

````python
def my_string_list_generator(m, n):
  """
  Creates m samples of n-length of random bases.
  Almost same as the one in Lab 3. But, it doesn't write to a file and it fills a line(list) of samples instead of lines in a file.
  Returns a list of random dna strings.
  m: number of samples per list/line.
  n: length of each sample.
  """

  # bases in a list
  chars=['A','T','G','C']

  # initialize a list to store random dna samples
  random_strings = []

  # for each element of the list/line
  for e in range(m):
    # create a empty line
    randomLine=''

    # for each position of each element of the list/line
    for c in range(n):

      randomIndx=random.randint(0,len(chars)-1) # random indexes
      randomLetter=chars[randomIndx]  # take the bases based on random indexes
      randomLine+=randomLetter  # fill the line

    # add the line/list to a list
    random_strings.append(randomLine)

  # return the list/line
  return random_strings
````

````python
def my_string_lists_generator(samples, lengths):
  """
  Creates a list of lists of random dna strings of several lengths.
  Uses my_string_list_generator function over a list of lengths.

  samples: number of examples per line/length.
  length: list of lengths.
  """

  # initialize a list to store te random generated lists
  generated_strings_lists = []

  # for each length in the list of lengths
  for l in lengths:
    # generate a random list of 'rows' samples and 'l' length for each one of them
    strings = my_string_list_generator(samples, l)
    #print(strings)
    generated_strings_lists.append(strings)

  # return the generated list of lists
  return generated_strings_lists
````

````python
def my_file_generator(strings_list, filename):
  """
  Creates a file with a list of lists of random dna strings of several lengths.
  """

  with open(filename, 'w') as file:
    for item in strings_list:
      file.write(str(item))
      file.write('\n')
````

Next, we calculate the Levenshtein distance for all paired dna strings of two files. To do this, we use some more simple function.

````python
def measure_levenshtein_distance(string1, string2):
    """
    Measures the Levenshtein distance between two DNA strings and their time of execution.
    """

    start_time = time.perf_counter()
    distance = Levenshtein.distance(string1, string2) # calculate the Levenshtein distance by using the appropriate library
    end_time = time.perf_counter()

    # print the distance and the time of execution
    #print(f"The Levenshtein distance between '{string1}' and '{string2}' is {distance}.")
    #print(f"The execution time is {end_time - start_time: 0.7f} seconds.")

    # return the distance and the time of execution
    return distance, end_time - start_time
````

````python
def read_from_file(filename):
  """
  Reads a file and returns a list of strings.
  """
  with open(filename, 'r') as file:
    return [line for line in file]
````

````python
def calculate_average_execution_time(file1, file2, samples_per_length):
  """
  Calculates the average execution time for Levenshtein distance calculation of the two given files.
  """

  # read the given files
  list_of_strings1 = read_from_file(file1)
  list_of_strings2 = read_from_file(file2)

  # initialize lists to store distances and average execution times
  distances = []
  execution_times = []
  avg_execution_times = []

  # loop over the lines
  for str1, str2 in zip(list_of_strings1, list_of_strings2):
    #print(f"String from file1: {str1}")
    #print(f"String from file2: {str2}")

    # calculate the Levenshtein distance for each line
    distance, execution_time = measure_levenshtein_distance(str1, str2)
    print(f"The Levenshtein distance between '{str1}' and '{str2}' is {distance}.")
    print(f"The execution time is {execution_time: 0.7f} seconds.")

    # update the lists for distances and times
    distances.append(distance)
    execution_times.append(execution_time)

    # calculate and print the average execution time of a line/list
    avg_execution_time = execution_time / samples_per_length
    print(f"The average time of calculation of the line is {avg_execution_time: 0.7f} seconds.\n")

    # update the average execution time list
    avg_execution_times.append(avg_execution_time)

  # return average execution times
  return avg_execution_times
````

````python
# calculate the average time for 5 samples per length
y_avg_execution_times = calculate_average_execution_time('file1.txt', 'file2.txt', samples_per_length)
````

Finally, we use the below simple function to plot the Levenshtein Distance Performance for the two above generated files.

````python
def plot_performance(x_string_lengths, y_avg_execution_times):
  """
  Plots a graph to show the performance: AVG execution (Y-axis) vs. string length (X-axis).
  """

  # create a figure for the plot
  plt.figure(figsize=(10, 6))
  # plot a line graph
  plt.plot(x_string_lengths, y_avg_execution_times, marker='o')

  # limit the x-axis
  #plt.xlim(x_string_lengths[0], x_string_lengths[-1])

  # adding labels and title
  plt.xlabel('String length')
  plt.ylabel('Average execution time')
  plt.title('Performance of Levenshtein distance calculation')

  # display the plot
  plt.grid(True) # add some grid in background
  plt.show()
````

````python
# plot the performance depending on the length of dna strings
plot_performance(lengths, y_avg_execution_times)
````

As we clearly observe, the average time of execution of calculating the Levenshtein distance generally increases as the length of strings increases. This is very reasonable as the algorithm needs to compare more characters(bases) per string per line, so it needs to perform more iterations(= more calculations + more time).
